```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv('all_seasons.csv')
df = df.drop(['Unnamed: 0'], axis=1)
df.head()
```

```python
dfnew = df[['player_name','age','player_height','player_weight','pts','reb']]
dfnew.head()
```

```python
dfnew['Target'] = 0
dfnew['Target'][(df['age']<=25)&(df['player_height']>=180)&(df['player_weight']<=90)&(df['pts']>=6)&(df['reb']>=3)] = 1

dfnew['Target_name'] = 'Non-Target'
dfnew['Target_name'][(df['age']<=25)&(df['player_height']>=180)&(df['player_weight']<=90)&(df['pts']>=6)&(df['reb']>=3)] = 'Target'

dfnew.head()
```

```python
X = dfnew.loc[:,['age','player_height','player_weight','pts','reb']]
Y = dfnew['Target']

from sklearn.model_selection import train_test_split
xtr , xts, ytr, yts = train_test_split(X, Y,test_size = .20, random_state=7649)
```

```python
from sklearn.linear_model import LogisticRegression
model_log = LogisticRegression()
model_log.fit(xtr,ytr)
prediksi_log = model_log.predict(xts)
skor_log = model_log.score(xts, yts)

print ("Logistic Regression Accuracy : %s" % "{0:.4%}".format(skor_log))
```

```python
from sklearn.ensemble import RandomForestClassifier
model_rf = RandomForestClassifier(n_estimators= 50)
model_rf.fit(xtr,ytr)
prediksi_rf = model_rf.predict(xts)
skor_rf = model_rf.score(xts, yts)

print ("Random Forest Accuracy : %s" % "{0:.4%}".format(skor_rf))
```

```python
from sklearn.neighbors import KNeighborsClassifier
model_knn = KNeighborsClassifier()
model_knn.fit(xtr,ytr)
prediksi_knn = model_knn.predict(xts)
skor_knn = model_knn.score(xts, yts)

print ("KNN Accuracy : %s" % "{0:.4%}".format(skor_knn))
```
